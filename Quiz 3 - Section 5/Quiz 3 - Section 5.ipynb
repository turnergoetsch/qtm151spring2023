(b) Recode a numeric colum

<font size = "3">

- Import the "data_raw/races.csv" file
- Recode the "year" column into a new column "year_brackets" <br>
with the following categories

$\qquad$ ``` ["1950-1974","1975-1999","2000-onwards"] ```

$\qquad$ HINT: Use the "pd.cut()" command.

````python
# Write your own code

races = pd.read_csv("data_raw/races.csv")

bins_labels = ["1950-1974", "1975-1999", "2000-onwards"]
bins_years = [1949, 1974, 1999, 2100]

races["year_brackets"] = pd.cut(races["year"],
                                bins = bins_years,
                                right = True,
                                labels = bins_labels)

races["year_brackets"]
````

(c) Aggregate and query

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Subset the observations for 'constructorId == 3' using ```.query()```
- Obtain the sum of "laps" by "driverId" <br> 
 using ``` .groupby().agg() ```
- Subset the drivers with a sum of laps $\ge 40$
- Store the results of all these operations in a dataframe <br>
```sprint_results_agg ```

````python
# Write your own code
sprint_results1 = pd.read_csv("data_raw/sprint_results.csv")

sprint_results_sums = (sprint_results1.query('constructorId == 3')
                      .groupby("driverId")
                      .agg(laps_sum = ('laps', 'sum')))

sprint_results_agg = sprint_results_sums.query('laps_sum >= 40')

sprint_results_agg
````

(d) Aggregate and sort

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Obtain the sum of "laps" by "driverId" <br> 
using ``` .groupby().agg() ```
- Sort the teams in descending order using of <br>
the sum of laps using ``` .sort_values() ```

````python
# Write your own code
sprint_results2 = pd.read_csv("data_raw/sprint_results.csv")

sprint_results_agg2 = (sprint_results2.groupby("driverId")
                  .agg(laps_sum = ("laps", 'sum'))
                  .sort_values('laps_sum', ascending= False))
````
